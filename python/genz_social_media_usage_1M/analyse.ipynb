```python
# Importation des libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
# Chargement du Dataframe

df = pd.read_csv('genz_social_media_usage_1M.csv', sep=',')
```

```python
# Les valeur unique de la variable "night_usage"

print(df['night_usage'].unique())
```

```python
# Changement du type de la variable "night_usage"

df['night_usage'] = df['night_usage'].astype(bool)
```

```python
# Vérification du type modifier

df['night_usage'].dtypes
```

```python
# Voir maintenant les valeurs unique du type modifier

print(df['night_usage'].unique())
```

```python
# Suppression de le variable "num_platforms_used"

df.drop('num_platforms_used', axis=1, inplace=True)
```

```python
# Création d'une nouvelle variable "category_age" (Pour séparer les "majeur" des "mineurs")

df['category_age'] = np.where(df['age'] >= 18, "Major", "Minor")
```

# Analyse univariée

```python
# Histogramme des variables "daily_usage_hours", "avg_session_minutes", "screen_time_before_sleep" et "mental_health_score"

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.histplot(data=df, x='daily_usage_hours', bins=30, kde=True)
plt.title('Histogramme de daily_usage_hours')
plt.ylabel('Fréquence')

plt.subplot(2, 2, 2)
sns.histplot(data=df, x='avg_session_minutes', bins=30, kde=True)
plt.title('Histogramme de avg_session_minutes')
plt.ylabel('Fréquence')

plt.subplot(2, 2, 3)
sns.histplot(data=df, x='screen_time_before_sleep', bins=30, kde=True)
plt.title('Histogramme de screen_time_before_sleep')
plt.ylabel('Fréquence')

plt.subplot(2, 2, 4)
sns.histplot(data=df, x='mental_health_score', bins=30, kde=True)
plt.title('Histogramme de mental_health_score')
plt.ylabel('Fréquence')

plt.tight_layout()
plt.show()
```

```python
# Nombre d'utilisateur par Plateforme

plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df, x='primary_platform', order=df['primary_platform'].value_counts(ascending=False).index)
ax.bar_label(ax.containers[0], padding=3)
plt.title('Répartition des utilisateurs par plateforme', fontsize='16', fontweight='bold', pad=20)
plt.xlabel('Plateforme', fontsize='15', fontweight='bold', labelpad=14)
plt.ylabel('Effectif', fontsize='15', fontweight='bold', labelpad=14)
plt.show()
```

```python
# Nombre d'utilisateur par pays

plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df, x='country', order=df['country'].value_counts(ascending=False).index)
ax.bar_label(ax.containers[0], padding=3)
plt.title('Répartition des utilisateurs par pays', fontsize='16', fontweight='bold', pad=20)
plt.xlabel('Pays', fontsize='15', fontweight='bold', labelpad=14)
plt.ylabel('Effectif', fontsize='15', fontweight='bold', labelpad=14)
plt.show()
```

```python
# Nombre d'utilisateur par niveau de connexion

plt.figure(figsize=(6, 5))
sns.countplot(data=df, x='addiction_level', order=df['addiction_level'].value_counts(ascending=False).index)
plt.title("Répartition des utilisateurs par intensité d'utilisation", fontsize='16', fontweight='bold', pad=20)
plt.xlabel("Niveau d'addiction", fontsize='15', fontweight='bold', labelpad=14)
plt.ylabel('Effectif', fontsize='15', fontweight='bold', labelpad=16)
plt.show()
```

```python
# Nombre d'utilisateur par raison d'utilisation

plt.figure(figsize=(10, 5))
sns.countplot(data=df, y='purpose', order=df['purpose'].value_counts(ascending=False).index)
plt.title("Répartition des utilisateurs par raison d'usage", fontsize='16', fontweight='bold', pad=20)
plt.xlabel('Effectif', fontsize='15', fontweight='bold', labelpad=14)
plt.ylabel("Raison usage", fontsize='15', fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.countplot(data=df, x='night_usage')
plt.title("Répartition des utilisateurs la nuit", fontsize='16', fontweight='bold', pad=20)
plt.xlabel('Utilisation la nuit', fontsize='15', fontweight='bold', labelpad=14)
plt.ylabel('Effectifs', fontsize='15', fontweight='bold', labelpad=14)
plt.show()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 2)
gender = df['gender'].value_counts()
label = ['Femme', 'Homme', 'Autre']
color = ['#4BCC63', '#FD3CD4', '#00FFFF']
plt.pie(gender, autopct='%1.2f%%', startangle=6, labeldistance=1.2, textprops={'fontsize': 14, 'weight': 'bold'}, colors=color)
plt.title("Répartition par genre", fontsize='16', fontweight='bold', pad=20)
plt.legend(label, title='Genre', title_fontsize=12, fontsize='large', loc='center right', bbox_to_anchor=(1, 0, 0.3, 1))

plt.subplot(1, 2, 1)
categorie_age = df['category_age'].value_counts()
label = ['Majeur', 'Mineur']
plt.pie(categorie_age, autopct='%1.2f%%', labeldistance=1.2, textprops={'fontsize': 14, 'weight': 'bold'})
plt.title("Répartition par categorie d'age", fontsize='16', fontweight='bold', pad=20)
plt.legend(label, title='Categorie age', title_fontsize=12, fontsize='large', loc='center right', bbox_to_anchor=(1, 0, 0.3, 1))

plt.tight_layout()
plt.show()
```

# Analyse multivarié

```python
plt.figure(figsize=(14, 16))

plt.subplot(2, 2, 1)
sns.boxplot(data=df, x='night_usage', y='mental_health_score', hue='night_usage', palette='deep', legend=False)
plt.title('mental_health_score by night_usage', fontsize='16', fontweight='bold', pad=20)

plt.subplot(2, 2, 2)
sns.boxplot(data=df, x='gender', y='mental_health_score', hue='gender', palette='bright', legend=False)
plt.title('mental_health_score par Genre', fontsize='16', fontweight='bold', pad=20)

plt.subplots_adjust(wspace=0.2, hspace=0.6)
plt.show()
```

```python
df_category_age = df[df['category_age'] == 'Minor']
df_category_age.groupby('addiction_level')[['daily_usage_hours', 'mental_health_score']].mean().sort_values(ascending=False, by='daily_usage_hours')
```

```python
df_category_age = df[df['category_age'] == 'Major']
df_category_age.groupby('addiction_level')[['daily_usage_hours', 'mental_health_score']].mean().sort_values(ascending=False, by='daily_usage_hours')
```

```python
# Calcul de la corrélation entre les variables numérique

corr_max = df.drop(columns=['gender', 'country', 'primary_platform', 'purpose', 'addiction_level', 'category_age', 'night_usage']).corr()
```

```python
# Graphique montrant les corrélations entre les variables

plt.figure(figsize=(10, 6))
sns.heatmap(corr_max, annot=True, cmap='coolwarm')
plt.title("Matrix de corrélation", fontsize='16', fontweight='bold', pad=20)
plt.xticks(rotation=40)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='daily_usage_hours', y='mental_health_score')
plt.title("Relation entre 'daily_usage_hours' \net 'mental_health_score'", fontsize='16', fontweight='bold', pad=20)
plt.show()
```
